### Davis Southern Club Women

These data were collected by Davis in the 1930s. They represent observed attendance at 14 social events by 18 Southern women. The result is a person-by-event matrix: cell (i,j) is 1 if person i attended social event j, and 0 otherwise.

REFERENCES

*    Breiger R. (1974). The duality of persons and groups. Social Forces, 53, 181-190.
*    Davis, A et al. (1941). Deep South. Chicago: University of Chicago Press.

### Import data

```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import networkx.algorithms.bipartite as bipartite

%matplotlib inline
plt.rcParams["figure.figsize"] = (12,10)

G = nx.davis_southern_women_graph()
women = G.graph['top']
clubs = G.graph['bottom']
```

```python
print(bipartite.biadjacency_matrix(G, women, clubs))
```

### Women
#### Women Degrees

```python
# project bipartite graph onto women nodes
W = bipartite.projected_graph(G, women)
print('')
print("# of Friends, Member")

W_degrees = {}
for w in women:
    W_degrees[w] = W.degree(w)

for w in sorted(W_degrees, key=W_degrees.get, reverse=True):
  print(W_degrees[w], w)
```

#### Women weights

```python
# project bipartite graph onto women nodes keeping number of co-occurence
# the degree computed is weighted and counts the total number of shared contacts
W = bipartite.weighted_projected_graph(G, women)
print('')
print("#Friend meetings, Member")

W_weights = {}
for w in women:
    W_weights[w] = W.degree(w, weight='weight')

for w in sorted(W_weights, key=W_weights.get, reverse=True):
    print(W_weights[w], w)
```


We can see a big difference betwwen the strength of connections when we evaluate simply the degrees vs the weights. Looking onnly at degrees it seems as if almost all the women are almost equally connected with only 3 women (Olivia Carleton, Flora Price, Charlotte McDowd) who have considerably less connections than the rest.

However when we look at the weighted connections, we can see that there are two women (Theresa Anderson and Evelyn Jefferson) who have considerably more influence than the rest and a more varying range of influence from top to bottom. Olivia Carleton and Flora Price are still considerably less influential, but Charlotte McDowd has more influence in her relationships despite the lower number of connectins.


#### Women Network Graphs

```python
weights=[edata['weight'] for f,t,edata in W.edges(data=True)]

nx.draw(W, with_labels=True, node_size=2000, node_color="orchid", font_size=12, font_family="calibri",
                width=weights, edge_color="darkgray", alpha=0.8)
```

```python
nx.draw_circular(W,width=weights, edge_color="darkgray", node_color="orchid", with_labels=True, node_size=2000, font_size=12, font_family="calibri",
                alpha=0.8)
```

The Circular layout shows two clear zones of influence or 'cliques'. One centered around the top right corner near Evelyn, Laura, Theresa and Brenda and the other around the bottom left near Katherina, Sylvia and Nora. While the Spring layout graph above clearly showed who were the most influential people, the two separate cliques were not readily apparent in that layout.

### Clubs
#### Club Degrees

```python
# project bipartite graph onto club nodes
C = bipartite.projected_graph(G, clubs)
print('')
print("#Clubs")

C_degrees = {}
for c in clubs:
    C_degrees[c] = C.degree(c)

for c in sorted(C_degrees, key=C_degrees.get, reverse=True):
  print(C_degrees[c], c)
```

Four clubs (E6, E7, E8, and E9) clearly stand out as more influential than the rest when looking only at degrees or number of connnections.

#### Club Weights

```python
# project bipartite graph onto women nodes keeping number of co-occurence
# the degree computed is weighted and counts the total number of shared contacts
C1 = bipartite.weighted_projected_graph(G, clubs)
print('')
print("#Club attendance, Clubs")

C_weights = {}
for c in clubs:
    C_weights[c] = C1.degree(c, weight='weight')

for c in sorted(C_weights, key=C_weights.get, reverse=True):
  print(C_weights[c], c)
```

Those same four clubs (E6, E7, E8, and E9) still stand out above the rest when looking at the weighted inflfuence, but E8 clearly stands far above all others including the other top three when evaluated by weight.

#### Club Network Graph

```python
weights=[edata['weight'] for f,t,edata in C1.edges(data=True)]

nx.draw(C,with_labels=True, node_size=1000, node_color="cornflowerblue", font_size=12, font_family="calibri",
                width=weights, edge_color="darkgray", alpha=0.8)
```

In the Spring layout graph you can clearly see the four most influential clubs (E6, E7, E8, and E9) at the center of the 'bowtie' shape and we can also see that E5, E3 and E12 are not far behind them.

```python
nx.draw_circular(C1, width=weights, edge_color="darkgray", node_color="cornflowerblue", with_labels=True, node_size=1000, font_size=12, font_family="calibri",
                alpha=0.8)
```

The Circular layout makes missing connections easier to spot for instance between E1 and E10 through E14 and between E14 and E1 through E5.
